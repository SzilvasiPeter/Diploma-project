```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
student_data = pd.read_csv('../../datasets/student-mat.csv')
```

```python
student_data.head()
```

```python
student_data['G1'].describe()
```

```python
categories = ['First Period', 'Second Period', 'Final']
student_score = [student_data['G1'].head(5),
                 student_data['G2'].head(5),
                 student_data['G3'].head(5)]

plt.plot(categories, student_score)
plt.title('Score Line Chart')
plt.xlabel('Periods')
plt.ylabel('Scores')
plt.show()
```
